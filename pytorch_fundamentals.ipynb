**Tensor Datatype**: Default is float32. 16 bit (half precision) floating point is faster but lower precision.

```python
# Float16 tensor
float_16_tensor = torch.tensor([3.0, 6.0, 9.0], dtype=torch.float16, device="cpu", requires_grad=False)
print(float_16_tensor.dtype)

# Float32 tensor
float_32_tensor = torch.tensor([3.0, 6.0, 9.0], dtype=torch.float32, device="cpu", requires_grad=False)
print(float_32_tensor.dtype)

# Not a good practice - Multiplying Float16 tensor with Float32 tensor
print((float_16_tensor*float_32_tensor).dtype)

# in32 tensor multiplication with float32 tesor
int_32_tensor = torch.tensor([3, 6, 9], dtype=torch.int32, device="cpu", requires_grad=False)
float_64_tensor = torch.tensor([3, 6, 9], dtype=torch.float64, device="cpu", requires_grad=False)
print((int_32_tensor*float_64_tensor).dtype)
print(int_32_tensor.device)
```

**Manipulating Tensors - Matrix Multiplication/ Dot Product**

```python
tensor_a = torch.randint(low=0, high =10, size=(2,3), dtype=torch.int32)
tensor_b = torch.randint(low=0, high =10, size=(3,2),  dtype=torch.int32)
tensor_a, tensor_b
```

```python
# Matrix multiplication by hand, i.e using for loops
%%time
result = torch.zeros((tensor_a.shape[0], tensor_b.shape[1]),dtype=torch.int32)
for i in range(tensor_a.shape[0]):      # rows of A
    for j in range(tensor_b.shape[1]):  # cols of B
        value = 0
        for k in range(tensor_a.shape[1]):  # shared dimension
          value += tensor_a[i][k] * tensor_b[k][j]
        result[i][j] = value
result
```

```python
# Matrix multiplication using torch.matmul()
%%time
torch.matmul(tensor_a, tensor_b)
```

```python
%%time
tensor_a @ tensor_b
```

**Reshaping, Stacking, Squeezing and Unsqueezing**

```python
import torch
x = torch.arange(1., 13.)
print(f"Original Tensor:{x} \nOriginal shape of x: {x.shape}")


# Change the shape with Reshape
  x_reshaped = torch.reshape(x , (2,6))
  print(f"x_reshaped: \n {x_reshaped},\nx_reshaped.shape: {x_reshaped.shape}")
  x_reshaped = torch.reshape(x , (6,2))
  print(f"x_reshaped: \n {x_reshaped},\nx_reshaped.shape: {x_reshaped.shape}")
  x_reshaped = torch.reshape(x , (3,4))
  print(f"x_reshaped: \n {x_reshaped},\nx_reshaped.shape: {x_reshaped.shape}")
  x_reshaped = torch.reshape(x , (4,3))
  print(f"x_reshaped: \n {x_reshaped},\nx_reshaped.shape: {x_reshaped.shape}")


```

```python
#  Stack tensors on top of each other
tensor_no = 2

if tensor_no == 1:
  x = torch.arange(1., 13.)
  print(f"Original Tensor:\n {x}\n")
  # Stacking with dim=0
  x_stacked = torch.stack([x, x, x, x], dim=0)
  print(f"Stacking with dim=0:\n {x_stacked.shape}\n {x_stacked}\n")

  # Stacking with dim=1
  x_stacked = torch.stack([x, x, x, x], dim=1)
  print(f"Stacking with dim=1:\n {x_stacked.shape}\n {x_stacked}")

if tensor_no == 2:
  tensor_2d = torch.tensor([[1, 2, 3], [4, 5, 6]])

  print(f"Original 2D tensor shape: {tensor_2d.shape}")
  print(f"Original 2D Tensor:\n {tensor_2d}")

  # Stacking with dim=0
  tensor_2d_stacked = torch.stack([tensor_2d, tensor_2d, tensor_2d, tensor_2d], dim=0)
  print(f"\nStacking with dim=0:\n {tensor_2d_stacked.shape}\n {tensor_2d_stacked}")

  # Stacking with dim=1
  tensor_2d_stacked = torch.stack([tensor_2d, tensor_2d, tensor_2d, tensor_2d], dim=1)
  print(f"\nStacking with dim=1:\n {tensor_2d_stacked.shape}\n {tensor_2d_stacked}")

  # Stacking with dim=2
  tensor_2d_stacked = torch.stack([tensor_2d, tensor_2d, tensor_2d, tensor_2d], dim=2)
  print(f"\nStacking with dim=2:\n {tensor_2d_stacked.shape}\n {tensor_2d_stacked}")

  # NOTE: which ever dim is chosen. that dim will have number of tensor stacked.
  # For eg, if dim = 0, No of times stacked = 4 --> (4, x, y)
  # For eg, if dim = 1, No of times stacked = 4 --> (x, 4, y)
  # For eg, if dim = 2, No of times stacked = 4 --> (x, y, 4)

```

```python
# Squeezing Tensors Removes single dimensions from the tensor

tensor_to_squeeze = torch.tensor([[[1, 2, 3], [4, 5, 6]]])
print(f"\ntensor_to_squeeze shape: {tensor_to_squeeze.shape}")
print(f"tensor_to_squeeze :\n {tensor_to_squeeze}")

squeezed_tensor = tensor_to_squeeze.squeeze(dim=0)
print(f"\nsqueezed_tensor shape: {squeezed_tensor.shape}")
print(f"squeezed_tensor :\n {squeezed_tensor}")

unsqueezed_tensor = torch.unsqueeze(squeezed_tensor, dim=2)
print(f"\nsqueezed_tensor shape: {unsqueezed_tensor.shape}")
print(f"squeezed_tensor :\n {unsqueezed_tensor}")
```

```python
#  Permute - used to swap dims and its a view, which means memory of the
#            tensor is the same and manipulations happen on the original tensor

x_original = torch.rand(size=(224, 224, 3)) # [height, width, colour_channels]
x_permuted = x_original.permute(2, 0, 1)  # [colour_channels, height, width]
print(f"Original shape: {x_original.shape}")
print(f"Permuted shape: {x_permuted.shape}")
```

**Numpy and Pytorch Tensors**

**Using seed: helps us get random tensors but the tensor is reproducible**

```python
import torch
RANDOM_SEED = 42

torch.manual_seed(RANDOM_SEED)
random_tensor_C = torch.rand(size=(3,4))
random_tensor_D = torch.rand(size=(3,4))

print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C == random_tensor_D)

print("\nAlways used torch.manual_seed(RANDOM_SEED) before definfing a tensor\n")

torch.manual_seed(RANDOM_SEED)
random_tensor_C = torch.rand(size=(3,4))

torch.manual_seed(RANDOM_SEED)
random_tensor_D = torch.rand(size=(3,4))

print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C == random_tensor_D)
```
